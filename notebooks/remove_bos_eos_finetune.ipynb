```python
import json
import re
input_path = '/content/drive/MyDrive/GourmetGPT/Dataset/structured_recipes_finetune.jsonl'
output_path = '/content/drive/MyDrive/GourmetGPT/Dataset/structured_recipes_finetune_nobos_eos.jsonl'

def remove_bos_eos(text):
    text = re.sub(r'^\[BOS\]\s*', '', text.strip())
    text = re.sub(r'\s*\[EOS\]$', '', text)
    return text.strip()

removed_count = 0
with open(input_path, 'r', encoding='utf-8') as fin, open(output_path, 'w', encoding='utf-8') as fout:
    for line in fin:
        obj = json.loads(line)
        original = obj['response']
        fixed = remove_bos_eos(original)
        if fixed != original:
            removed_count += 1
        obj['response'] = fixed
        fout.write(json.dumps(obj, ensure_ascii=False) + '
')
print(f'Removed [BOS] and [EOS] from {removed_count} entries. Output written to {output_path}')
```
